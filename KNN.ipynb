```python
import pandas as pd
```

```python
data=pd.read_csv("student_knn_classification.csv")
```

```python
data.head()
```

```python
from sklearn.neighbors import KNeighborsClassifier
x=data[["Study_Hours","Attendance"]]
y=data["Result"]
model=KNeighborsClassifier()
model.fit(x,y)
new_data=pd.DataFrame({
    "Study_Hours":[7],
    "Attendance" : [80]
})
res=model.predict(new_data)
print(res)
```
