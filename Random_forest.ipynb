```python
#----------Random Forest--------classification--------------#
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
```

```python
data=pd.read_csv("student_knn_classification.csv")
```

```python
data.head()
```

```python
x=data[["Study_Hours","Attendance"]]
y=data["Result"]
model=RandomForestClassifier(n_estimators=4)
model.fit(x,y)
new_data=pd.DataFrame({
    "Study_Hours":[1],
    "Attendance":[35]
})
res=model.predict(new_data)
print(res)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
data1=pd.read_csv("student_knn_regression.csv")
```

```python
data1.head()
```

```python
x=data1[["Study_Hours","Attendance"]]
y=data1["Marks"]
model = RandomForestRegressor(n_estimators=4)
model.fit(x,y)
new_data=pd.DataFrame({
    "Study_Hours":[1],
    "Attendance":[35]
})
res=model.predict(new_data)
print(res)
```
